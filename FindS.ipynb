```python
attribute = [['Sunny','Rainy'],
['Warm','Cold'],
['Normal','High'],
['Strong','Weak'],
['Warm','Cool'],
['Same','Change']]

num_attr = len(attribute)
print (" \n The most general hypothesis : ['?','?','?','?','?','?']\n")
print ("\n The most specific hypothesis : ['Phi','Phi','Phi','Phi','Phi','Phi']\n")
a = []
print("\n The Given Training Data Set \n")
with open('/content/gdrive/My Drive/ML LAB/iris.csv', 'r') as csvFile:
  reader = csv.reader(csvFile) 
  for row in reader:
    a.append (row) 
    print(row)

print("The initial value of hypothesis: ")
hypothesis = ['0'] * num_attr
print(hypothesis)


```

```python
for j in range(0,num_attr):
  hypothesis[j] = a[0][j];
  print(a[0][j])
print("\n Find S: Finding a Maximally Specific Hypothesis\n")
for i in range(0,len(a)):
  if a[i][num_attr]=='Yes':
    for j in range(0,num_attr):
      if a[i][j]!=hypothesis[j]:
        hypothesis[j]='?'
      else :
        hypothesis[j]= a[i][j]
    print(" For Training Example No :{0} the hypothesis is ".format(i),hypothesis)
print("\n The Maximally Specific Hypothesis for a given Training Examples :\n")
print(hypothesis)
```
